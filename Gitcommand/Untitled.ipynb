ローカルリポジトリが外部へアップロードする領域

リモートリポジトリがGithub側のリポジトリです。

それぞれの領域にファイル情報を移動する際にコマンド入力で移動しますが、この時にどこにいてどのファイルをどのような操作をするかと言うことが重要になってきます。

そして、全ての根本的な情報として初期化された時の設定が引き継がれると言う事を忘れないようにしましょう。

例えば、１つのフォルダに２つのフォルダがあって、その片方をリモートリポジトリにプッシュした後に、もう片方を同じリモートリポジトリ内で入れ替える、と言った時に、入れ替えではなく後から入れ替えるフォルダが追加になってしまうことや、リモートリポジトリを別のものに変更して、そちらに別のフォルダをプッシュする時に、リモートリポジトリのURLがgitコマンド上で変更されておらず、先にプッシュしたリモートリポジトリの中に追加で入れてしまう、などと言ったことが発生します。

高度なコマンドを使いこなしてそれぞれの処理をすることも最終的には必要かもしれませんが、まずは構造を知って間違いのない処理ができるような手順を理解しましょう。

そのために知っておきたいことは

$ ls -a で隠しファイルが見れる

$ rm -rf .git で.gitフォルダを削除できる(git initの設定が削除される)

$ git init で再度設定をし直す

と言うことです。

ひとつひとつの処理で設定を最初から行えば、上記のような手違いが起こる心配がなく、また初学者の理解についてもより一層速い習得が見込めます。
